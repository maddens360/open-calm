```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained("cyberagent/open-calm-3b", device_map="auto", torch_dtype=torch.float16)
tokenizer = AutoTokenizer.from_pretrained("cyberagent/open-calm-3b")

```

```python
prompt = "今後の私たちの生活は、"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
with torch.no_grad():
    tokens = model.generate(
        **inputs,
        max_new_tokens=64,
        do_sample=True,
        temperature=0.7,
        pad_token_id=tokenizer.pad_token_id,
    )
    
output = tokenizer.decode(tokens[0], skip_special_tokens=True)
print(output)
```

```python
import urllib.request
import re
import numpy as np
import tensorflow_hub as hub
import gradio as gr
import os
```

```python
def generate(userPrompt):
  inputs = tokenizer(userPrompt, return_tensors="pt").to(model.device)
  with torch.no_grad():
      tokens = model.generate(
          **inputs,
          max_new_tokens=64,
          do_sample=True,
          temperature=0.7,
          pad_token_id=tokenizer.pad_token_id,
      )
      
  output = tokenizer.decode(tokens[0], skip_special_tokens=True)
  return output
```

```python

title = 'Open Calm 3b'

with gr.Blocks() as demo:

    gr.Markdown(f'<center><h1>{title}</h1></center>')

    with gr.Row():
        
        with gr.Group():
            prompt = gr.Textbox(label='プロンプトを入力してください')
            btn = gr.Button(value='Submit')
            btn.style(full_width=True)

        with gr.Group():
            answer = gr.Textbox(label='出力:')

        btn.click(generate, inputs=[prompt], outputs=[answer])
# demo.launch()
demo.launch(share=True)
```
